```python
import nltk
from nltk import ngrams
from collections import Counter
nltk.download('punkt')
text = """Massive landslides hit the district of Wayanad
 in northern Kerala in the early hours of July 30, with continuous torrential rainfall triggering large-
  death and destruction in Mundakkai and Chooralmala of Meppadi panchayat, with entire villages being washed away."""
tokens = nltk.word_tokenize(text)
trigrams = list(ngrams(tokens, 3))
trigram_freq = Counter(trigrams)
for trigram, freq in trigram_freq.items():
    print(f"{trigram}: {freq}")
```
